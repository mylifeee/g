```python
#8th program
import sqlite3
conn = sqlite3.connect("D:\\data wrangling\\New folder\\example.db")
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS example_table (
        id INTEGER PRIMARY KEY,
        name TEXT,
        age INTEGER
    )
''')
data_to_insert = [
    (1, 'Pranav', 25),
    (2, 'Ansari', 30),
    (3, 'Binu', 22)
]
cursor.executemany('''
    INSERT INTO example_table (id, name, age) VALUES (?, ?, ?)
''', data_to_insert)
conn.commit()
conn.close()

```

```python
from xml.dom import minidom
# parse an xml file by name
file = minidom.parse("D:\\data wrangling\\ex4.xml")

#use getElementsByTagName() to get tag
models = file.getElementsByTagName('model')

# one specific item attribute
print('model #2 attribute:')
print(models[1].attributes['name'].value)
# all item attributes
print('\nAll attributes:')
for elem in models:
  print(elem.attributes['name'].value)
# one specific item's data
print('\nmodel #2 data:')
print(models[1].firstChild.data)
print(models[1].childNodes[0].data)
# all items data
print('\nAll model data:')
for elem in models:
  print(elem.firstChild.data)

```
